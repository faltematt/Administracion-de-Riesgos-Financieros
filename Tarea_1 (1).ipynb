##Sector/Temática

Tanto en la actualidad como en los tiempos pasados, la minería ha desempeñado un rol de suma importancia en el desarrollo económico y social de los países. Específicamente en Chile, la minería de cobre representa una de las mayores actividades económicas, exportando aproximadamente un 30% de la producción global.

Ahora bien, durante los últimos años, el litio, gracias a sus propiedades químicas, ha ido tomando gran fuerza en la industria principalmente eléctrica. Este elemento metálico blando, de color blanco/plata, permite el almacenamiento de grandes densidades de energia eléctrica en espacios relativamente pequeños. Por lo que lo hace un elemento ideal para las baterías de, por ejemplo, autos eléctricos. 

En este sentido, el sector que se analizará en el presente trabajo será el minero, pero específicamente el del litio. Para esto se profundizará en una de las mayores compañías productoras de este metal, Soquimich (Sociedad Química y Minera de Chile), de ahora en adelante SQM.

En relación a lo anterior, el Gerente General de SQM, Ricardo Ramos expresó en el último informe de utilidades del segundo semestre de 2022 lo siguiente: “Estamos cercanos a llegar a las 180.000 toneladas métricas en nuestra capacidad de litio. Hoy estamos trabajando para completar una capacidad de carbonato de litio de 210.000 toneladas métricas de producto codiciado, de alta calidad y valor agregado que se producirá en Chile. Seguimos comprometidos a reducir nuestro uso de salmuera y agua a través de la tecnología y la innovación continua. Esta nueva capacidad nos permitirá producir productos de litio de alto valor agregado para la producción de más de 5 millones de vehículos eléctricos”.

##Derivados

Se llevó a cabo una valoracion de una opcion de compra de la accion de SQM transado en NYSE. El precio de ejercicio es 145 USD y la fecha de vencimiento es en enero de 2024.

```python
from sympy import *
init_printing()
```

```python
# Parámetros Bono
r, B = symbols('r B')
# Parámetros Acción
u, d, S0 = symbols('u d S0')
# Parámetros Bono
r, B = symbols('r B')
# Parámetros Acción
u, d, S0 = symbols('u d S0')
W = Matrix([[-B, -S0], [r*B, u*S0],[r*B, d*S0]])
W
pi_u, pi_d = symbols('pi_u pi_d')
pi = Matrix([[1],[pi_u],[pi_d]])
pi
NA=pi.transpose()*W
NA
# Solver tutorial
# https://docs.sympy.org/latest/modules/solvers/solvers.html
piNA=solve([NA[0],NA[1]],[pi_u,pi_d],dict=True)
piNA
```

```python

# Parámetros Opción
q3 = symbols('q3')
K = 145                       #strike
# Parámetros Bono
r, B = 1.0379, 98.98          #tasa libre de riesgoa dos años, valor par bono https://www.bloomberg.com/markets/rates-bonds/government-bonds/us
# Parámetros Acción
u, d, S0 = 1.5013, 0.4987, 113.52   #retorno favorable, retorno desfavorable (volatilidad implicada 50,13%) y precio del activo https://finance.yahoo.com/quote/SQM/options?date=1705622400&p=SQM
W = Matrix([[-B, -S0, -q3], [r*B, u*S0, max(0,u*S0-K)],[r*B, d*S0, max(0,d*S0-K)]])
W
```

```python
u,r,d = symbols('u r d')
Prices=Matrix([[1],[piNA[0][pi_u]],[piNA[0][pi_d]]]).transpose()*W
Prices
```

```python
Prices.subs([(d,0.4987),(u,1.5013),(r,1.0379)])
```

En consecuencia, el valor de la opcion de compra, con un strike de USD145 y vencimiento el 19 de enero de 2024, es de USD13.1756
